# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
import requests
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
print(btc)
# "data": {
#     "opening_price": "4269000",  최근 24시간 내 시작 거래금액
#     "closing_price": "4329000",  최근 24시간 내 마지막 거래금액
#     "min_price": "4240000",  최근 24시간 내 최저 거래금액
#     "max_price": "4379000",  최근 24시간 내 최고 거래금액
#     "average_price": "4308915.6601",
#     "units_traded": "15819.02481803",
#     "volume_1day": "15819.02481803",
#     "volume_7day": "43059.00396634",
#     "buy_price": "4330000",
#     "sell_price": "4334000",
#     "24H_fluctate": "60000",
#     "24H_fluctate_rate": "1.40",
#     "date": "1546490680295"
# 최고가와 최저가의 차이를 변동폭으로 정의할 때 
# (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.
````

````python
# 아래에 코드를 작성하세요.
open_price = int(btc["opening_price"]) #시가
max_price = int(btc["max_price"]) #최대가
min_price = int(btc["min_price"]) #최소가
fluctate = max_price - min_price # 변동폭
if open_price + fluctate > max_price:
    print('상승장')
else:
    print('하락장')
````

## 모든 가상화폐 시세 판단하기


> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.
>
> 예외 처리에 유의하세요.



```
예시 출력)
BTC 상승장
ETH 상승장
DASH 상승장
LTC 하락장
ETC 하락장
XRP 상승장
BCH 상승장
...
```


````python
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
print(currency)
````

````python
# 아래에 코드를 작성하세요.
"""> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.
>
> 예외 처리에 유의하세요.



```
예시 출력)
BTC 상승장
ETH 상승장
DASH 상승장
LTC 하락장
ETC 하락장
XRP 상승장
BCH 상승장
...
"""
#방법1
for key in currency.keys():
    if type(currency[key]) == dict: #예외처리   or if type(value) == type({}):
        open_price = float(currency[key]['opening_price'])
        max_price = float(currency[key]['max_price'])
        min_price = float(currency[key]['min_price'])
        fluctate = float(currency[key]['max_price']) - float(currency[key]['min_price'])
        if open_price + fluctate > max_price:
            print(f'{key} 상승장')
        else:
            print(f'{key} 하락장') 
print('\n')
#방법2
for key, value in currency.items(): #key == 'BTC'... value == {...}
    if 'max_price' in value: # 예외처리
        open_price = float(value['opening_price'])
        max_price = float(value['max_price'])
        min_price = float(value['min_price'])
        if open_price + (max_price - min_price) > max_price:
            print(f"{key} 상승장")
        else:
            print(f"{key} 하락장")
print('\n') 
#방법3
for key, value in currency.items(): #key == 'BTC'... value == {...}
    if isinstance(value, dict): # 예외처리
        open_price = float(value['opening_price'])
        max_price = float(value['max_price'])
        min_price = float(value['min_price'])
        if open_price + (max_price - min_price) > max_price:
            print(f"{key} 상승장")
        else:
            print(f"{key} 하락장")
print('\n')
#방법4
for key, value in currency.items():
    try:
        open_price = float(value['opening_price'])
        max_price = float(value['max_price'])
        min_price = float(value['min_price'])
        if open_price + (max_price - min_price) > max_price:
             print(f"{key} 상승장")
        else:
            print(f"{key} 하락장")
    except TypeError:
        print("타입 에러 발생")
    
    
    
````
